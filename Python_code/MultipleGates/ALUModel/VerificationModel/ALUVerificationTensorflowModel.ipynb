```python
def print_unique_occurrences(lst):
    # Count the occurrences of each value in the list
    counts = {}
    for value in lst:
        counts[value] = counts.get(value, 0) + 1

    # Sort the unique values by their occurrences in descending order
    unique_values = sorted(counts.keys(), key=lambda x: counts[x], reverse=True)

    # Print the unique values along with their occurrences
    for value in unique_values:
        print(f"{value}: {counts[value]} occurrences")
```
