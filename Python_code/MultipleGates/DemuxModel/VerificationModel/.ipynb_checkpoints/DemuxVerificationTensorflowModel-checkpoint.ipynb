```python
import tensorflow as tf
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import pickle
from matplotlib import pyplot
import collections
from keras.optimizers import SGD
np.set_printoptions(threshold=np.sys.maxsize)

def count_occurrences(arr):
    occurrence_counts = {}
    
    # Count the occurrences of each integer
    for num in arr:
        if num in occurrence_counts:
            occurrence_counts[num] += 1
        else:
            occurrence_counts[num] = 1
    
    # Group the numbers by occurrence count
    occurrence_groups = {}
    for num, count in occurrence_counts.items():
        if count in occurrence_groups:
            occurrence_groups[count].append(num)
        else:
            occurrence_groups[count] = [num]
    
    # Sort the occurrence groups by the occurrence count (in descending order)
    sorted_groups = sorted(occurrence_groups.items(), reverse=True)
    
    # Print the numbers and their occurrence counts
    for count, nums in sorted_groups:
        nums_str = ", ".join([str(num) for num in nums])
        print(f"{nums_str}: {count}")
```

```python
#define the data
FullData = pd.read_csv("Demux_Verification_Data.csv")

x, y = FullData.values[:, 2:-1], FullData.values[:, -1]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
n_features = x_train.shape[1]
```

```python
train_acc = 0
valuelist = [];
while(train_acc < 0.999):
    import tensorflow as tf
    import pandas as pd
    import numpy as np
    from sklearn.model_selection import train_test_split
    import pickle
    import collections
    from matplotlib import pyplot
    from keras.optimizers import SGD
    np.set_printoptions(threshold=np.sys.maxsize)

    #define the data
    FullData = pd.read_csv("Demux_Verification_Data.csv")

    x, y = FullData.values[:, 2:-1], FullData.values[:, -1]
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5)
    n_features = x_train.shape[1]

    # define the model
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Dense(100, input_shape=(n_features,), activation= "relu"))
    model.add(tf.keras.layers.Dense(80, activation= "relu"))
    model.add(tf.keras.layers.Dense(80, activation= "relu"))
    model.add(tf.keras.layers.Dense(80, activation= "relu"))
    model.add(tf.keras.layers.Dense(30, activation= "relu"))
    model.add(tf.keras.layers.Dense(30, activation= "relu"))
    model.add(tf.keras.layers.Dense(30, activation= "relu"))
    model.add(tf.keras.layers.Dense(30, activation= "relu"))
    model.add(tf.keras.layers.Dense(10, activation= "relu"))
    model.add(tf.keras.layers.Dense(10, activation= "relu"))
    model.add(tf.keras.layers.Dense(10, activation= "relu"))
    model.add(tf.keras.layers.Dense(5, activation= "relu"))
    model.add(tf.keras.layers.Dense(5, activation= "relu"))
    model.add(tf.keras.layers.Dense(5, activation= "relu"))
    model.add(tf.keras.layers.Dense(5, activation= "relu"))
    model.add(tf.keras.layers.Dense(5, activation= "relu"))
    model.add(tf.keras.layers.Dense(1, activation='sigmoid'))

    predictions = model(np.asarray(x_train).astype('float32')).numpy()
    opt = SGD(lr=0.01)


    model.compile(optimizer=opt, loss="binary_crossentropy", metrics=['accuracy'])

    model.fit(x_train, y_train, epochs=15, verbose = 1)
    
    _, train_acc = model.evaluate(x_train, y_train, verbose = 1)
    _, test_acc = model.evaluate(x_test, y_test, verbose=0)

    y_pred = model.predict(x_test)
    y_pred_classes = np.round(y_pred).flatten()

    # compute class-wise accuracy and confusion matrix
    class_0_correct = 0
    class_0_total = 0
    class_0_true_0 = 0
    class_0_false_1 = 0
    class_1_correct = 0
    class_1_total = 0
    class_1_true_1 = 0
    class_1_false_0 = 0
    for i in range(len(y_test)):
        true_class = y_test[i]
        pred_class = y_pred_classes[i]
        if true_class == 0:
            class_0_total += 1
            if pred_class == 0:
                class_0_correct += 1
                class_0_true_0 += 1
            else:
                class_0_false_1 += 1
                if((train_acc > 0.8)):
                    valuelist.append(str(i))
        elif true_class == 1:
            class_1_total += 1
            if pred_class == 1:
                class_1_correct += 1
                class_1_true_1 += 1
            else:
                class_1_false_0 += 1


    class_0_accuracy = class_0_correct / class_0_total
    class_1_accuracy = class_1_correct / class_1_total
    
    if(train_acc > 0.98):
        print(f"  True 0: {class_0_true_0}")
        print(f"  False 1: {class_0_false_1}")

        print(f"  True 1: {class_1_true_1}")
        print(f"  False 0: {class_1_false_0}")


history = model.fit(x_train, y_train, epochs=100)
_, train_acc = model.evaluate(x_train, y_train, verbose=0)
_, test_acc = model.evaluate(x_test, y_test, verbose=0)
print('Train: %.3f, Test: %.3f' % (train_acc, test_acc))
# plot loss during training
pyplot.subplot(211)
pyplot.title('Loss')
pyplot.plot(history.history['loss'], label='train')
pyplot.legend()
# plot accuracy during training
pyplot.subplot(212)
pyplot.title('Accuracy')
pyplot.plot(history.history['accuracy'], label='train')
pyplot.legend()
pyplot.show()

with open('Demuxverificationmodel.pkl', 'wb') as f:
    pickle.dump({'weights': model.get_weights(), 'history': history.history}, f)
```
